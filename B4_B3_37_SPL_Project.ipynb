```python
import pandas as pd

df = pd.read_csv("Churn_Modelling.csv")
df.head()
```

```python
df = df.drop(['RowNumber','CustomerId','Surname'], axis=1)
df = pd.get_dummies(df, drop_first=True)
df.head()
```

```python
X = df.drop('Exited', axis=1)
y = df['Exited']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
import gradio as gr
import pickle
import numpy as np

# Load model
model = pickle.load(open("model.pkl", "rb"))

# Prediction function
def predict(credit, age, tenure, balance, products, card, active, salary, germany, spain, male):
    try:
        input_data = np.array([[credit, age, tenure, balance, products, card, active, salary, germany, spain, male]])
        result = model.predict(input_data)

        if result[0] == 1:
            return "Customer will Churn"
        else:
            return "Customer will NOT Churn"
    except:
        return "Error in prediction"

# Interface
interface = gr.Interface(
    fn=predict,
    inputs=[
        gr.Number(label="Credit Score"),
        gr.Number(label="Age"),
        gr.Number(label="Tenure"),
        gr.Number(label="Balance"),
        gr.Number(label="Number of Products"),
        gr.Radio([0,1], label="Has Credit Card (0=No, 1=Yes)"),
        gr.Radio([0,1], label="Is Active Member (0=No, 1=Yes)"),
        gr.Number(label="Estimated Salary"),
        gr.Radio([0,1], label="Geography: Germany (1=Yes, 0=No)"),
        gr.Radio([0,1], label="Geography: Spain (1=Yes, 0=No)"),
        gr.Radio([0,1], label="Gender: Male (1=Yes, 0=No)")
    ],
    outputs="text",
    title="Bank Customer Churn Prediction"
)

interface.launch()
```
